```python
from sklearn.model_selection import train_test_split
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier
```

```python
iris_data = datasets.load_iris()
iris_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size = 0.25, random_state = 4)
x_train.shape
```

```python
rfc = RandomForestClassifier(n_estimators=50, max_features = 'auto')
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
rfc = RandomForestClassifier(n_estimators=20, max_features = 'auto')
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
rfc = RandomForestClassifier(n_estimators=2, max_features = 'auto')
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
rfc = RandomForestClassifier(n_estimators=50, max_features = 'log2')
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
house_data = datasets.load_boston()
house_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(house_data.data, house_data.target, test_size = 0.25, random_state = 4)
```

```python
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred = rfr.predict(x_test)
metrics.mean_squared_error(y_test, y_pred)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
metrics.mean_squared_error(y_test, y_pred)
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
metrics.mean_squared_error(y_test, y_pred)
```
